```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("snunlp/KR-SBERT-V40K-klueNLI-augSTS")

sent1 = "침실의 에어컨 온도를 24도로 맞춰줘."
sent2 = "침실 에어컨 온도 24도 설정."
embeddings = model.encode([sent1, sent2], convert_to_tensor=True)

similarity = util.pytorch_cos_sim(embeddings[0], embeddings[1]).item()
print(f"유사도: {similarity:.4f}")
```

```python
sentences = [
    "침실의 에어컨 온도를 24도로 맞춰줘.",  # 기준 문장
    "침실 에어컨 온도 24도 설정.",        # GPT B
    "매분마다 침실의 에어컨 온도를 24도로 설정해줘.",  # GPT C
    "방의 에어컨을 24도로 해줘."         # Gemini
]

model = SentenceTransformer("snunlp/KR-SBERT-V40K-klueNLI-augSTS")
embeddings = model.encode(sentences, convert_to_tensor=True)

for i, label in enumerate(["GPT B", "GPT C", "Gemini"], start=1):
    sim = util.pytorch_cos_sim(embeddings[0], embeddings[i]).item()
    print(f"{label} 유사도 (SBERT): {sim:.4f}")
```

# 핵심의도 중심
원래 문장 ~ 정규화 결과

"매분마다 침실의 에어컨 온도를 24도로 설정해줘." ~ "침실 에어컨 온도 24도 설정"

"침실의 에어컨 온도를 24도로 맞춰줘." ~ "침실 에어컨 온도 24도 설정"

# 키워드 매핑
정규화 + 토큰화 + 키워드 사전 기반 매핑

"매분마다" → when = 매분마다 와 같이 분리

그 외 단어도 디바이스/장소/값으로 맵핑

매분마다 침실의 에어컨 온도를 24도로 설정해줘. ~ {"location": "침실", "device": "에어컨", "action": "설정", "value": "24", "when": "매분마다"}
